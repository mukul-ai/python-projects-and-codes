```python
pin = 1234
def welcome():
    print("welcome to the SBI bank ATM")
    autho()
    result = True   

def autho():
    global pin
    flag=True 
    count=0
    global result
    while flag:
        upin=int(input("enter the ATM pin:"))
        if(pin==upin):
            print("you have successfully login")
            result=True
            menu(result)
            flag= False
            
        else:
            print("you have enter incorrect pin,plz try again")
            count=count+1
            if(count==3):
                print("you tried many times,your attempt is over and your card is blocked")
                cardblock(result)
                flag=False


#Balance = 10000     
#result = True
def menu(result): 
 #   global Balance
    global pin
    #global result
    Flag = True
    while Flag:
        print("opt = 1 for Balance check")
        
        opt=int(input("select option:"))
        if(opt==1):
            balance()
            autho()
        elif(opt==2):
            Deposite()
            autho()
        elif(opt==3):
            withdraw()
           # autho()
        elif(opt==4):
            pinchange()
            #autho()
        elif(opt==5):
            logout1()
            Flag = False

            

Balance = 10000            
def balance():
    global Balance
    print("Balance", Balance)

def Deposite():
    global Balance
    deposite=int(input("enter the deposite amount:"))
    Balance=Balance+deposite
    print("your account balance is:",Balance)
    
    
def withdraw():
    #withdraw =0
    global Balance
    withdraw=int(input("enter the withdraw amount:"))
    if(withdraw > Balance):
        print("enter the withdraw amount less than balance amount:")
    else:
        Balance=Balance-withdraw
        print("your remaining balance is:",Balance)
        
        
def pinchange():
    global pin
    flag =True
    while flag:
        new_pin=int(input("enter the new pin:"))
        if(pin==new_pin):
            print("your new pin is same as previous one,so plz enter the diffrent pin:")
        else:
            pin=new_pin
            print("your pin has been change successfully")
            flag = False
            
def logout1():
    print("you are successfully log out,plz collect your card")    
    Flag=False  
    flag=False  

    
            
def cardblock(result):
    result = False
    print("card Block")
    flag=False  
    
welcome()  
```
